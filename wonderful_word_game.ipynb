#### 유사한 단어 찾기 게임
1. 사전 학습된 모델 또는 적절한 데이터 셋을 찾는다
2. 워드 임베딩 모델을 학습시킨다
3. 단어 유사도가 0.8 이상인 A,B를 랜덤 추출한다
4. A, B와 대응되는 C를 추출한다
5. D를 입력받는다

=> <br>
A : B = C : D 관계에 대응하는 D를 찾는 게임을 만든다
<br>
ex) A : 산, B : 바다, C : 나무, D : 물 

**<출력 예시>**

관계 [수긍 : 추락 = 대사관 : ?] <br>
모델이 예측한 가장 적합한 단어 : 잠입 <br>
당신의 답변과 모델 예측의 유사도 : 0.34 <br>

```python
# ★ JSON들이 모여 있는 폴더 (Windows 경로는 r"..." 로 쓰면 안전)
IN_DIR = Path(r"C:\\skn_17\\nlp\\03_word_embedding\\TL1\\IT_과학")
OUT_PATH = Path(r"C:\\skn_17\\nlp\\03_word_embedding\\TL1\sentences_IT_과학.txt")

def iter_sentences(obj):
    """중첩된 dict/list 구조에서 'sentence' 키의 문자열을 모두 추출"""
    if isinstance(obj, dict):
        for k, v in obj.items():
            if k == "sentence" and isinstance(v, str):
                s = v.strip()
                if s:
                    yield s
            else:
                yield from iter_sentences(v)
    elif isinstance(obj, list):
        for item in obj:
            yield from iter_sentences(item)

count_files = 0
count_sentences = 0
all_sentences_it = []
# **/*.json : 하위 폴더까지 재귀적으로 탐색
for fp in sorted(IN_DIR.glob("**/*.json")):
    try:
        with fp.open("r", encoding="utf-8") as f:
            data = json.load(f)
        if count_files == 800:
            break
        extracted = list(iter_sentences(data))
        all_sentences_it.extend(extracted)
        count_files += 1
        count_sentences += len(extracted)
    except Exception as e:
        print(f"[WARN] {fp} 처리 실패: {e}")

# 한 줄에 문장 하나씩 저장
OUT_PATH.parent.mkdir(parents=True, exist_ok=True)
with OUT_PATH.open("w", encoding="utf-8") as out:
    for s in all_sentences_it:
        out.write(s + "\n")

print(f"[완료] 파일 {count_files}개에서 문장 {count_sentences}개 추출 → {OUT_PATH}")
```

```python
def start_game():
    while True:
        # 랜덤 단어 뽑기
        random_word = random.choice(loaded.wv.index_to_key)
        
        # 랜덤 단어와 유사도가 0.8 이상인 단어 뽑기 
        similar_wordlist = loaded.wv.most_similar(random_word)
        similar_words = [word for word, sim in similar_wordlist if sim >= 0.8]

        if similar_words:  
            break
    
    b = random.choice(similar_words)
    a_vec =  loaded.wv[random_word]
    # c는 랜덤 단어와 반대의 벡터
    c, _ = loaded.wv.most_similar(-a_vec, topn=1)[0]

    # 관계 계산 (벡터 연산)
    target = loaded.wv[c] + (loaded.wv[b] - loaded.wv[random_word])
    cands = loaded.wv.similar_by_vector(target, topn=50)  # 넉넉히 뽑기

    ban = {c, b, random_word}
    d = next(((w, sim) for w, sim in cands if w not in ban), None)

    print(f'관계 [{random_word} : {b} = {c} : ?]')
    answer = input('D를 입력하세요 : ').strip()

    print(f'모델이 예측한 가장 적합한 단어 {d}')
    print(f'당신의 답변과 모델 예측의 유사도 {loaded.wv.similarity(c, answer)}')

    if d == answer:
        print('잘 하셨습니다!')
    else:
        print('아쉽네요, 더 생각해보세요')
```

```python
start_game()
```
